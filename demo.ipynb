```python
import cv2
import util
import numpy as np
import matplotlib.pyplot as plt
from car_detectors import MobileNetSSDCarDetector
from position_predictor import PositionPredictor
```

## Car Detector with pre-trained MobileNet SSD

```python
car_detector = MobileNetSSDCarDetector()
```

```python
img_id = 4
folder = 'train'
img = util.read_velocity_image(folder, img_id, False)
```

```python
plt.imshow(img)
```

```python
bboxes = car_detector.predict(img)
res = util.add_bounding_boxes(img, bboxes)
plt.imshow(res)
```

## Predict Positions

```python
pp = PositionPredictor()
```

```python
# predicted positions
img = pp.predict(img, bboxes)
fig = plt.figure(figsize=(15,15))
plt.imshow(img)
```

```python
# ground truth positions
img = util.read_velocity_image(folder, img_id, True)
fig = plt.figure(figsize=(15,15))
plt.imshow(img)
```
