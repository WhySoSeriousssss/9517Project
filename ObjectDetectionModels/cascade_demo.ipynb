# Car detector based on CascadeClassifier


Ref: https://dev.to/kalebu/how-to-perform-real-time-vehicle-detection-in-python-4i9h


```python
import cv2
import util
import matplotlib.pyplot as plt
```

```python
cars_cascade = cv2.CascadeClassifier('haarcascade_car.xml')
```

```python
image = util.read_velocity_image('supp', 8, annotation=True)
plt.imshow(image)
```

```python
cars = cars_cascade.detectMultiScale(image, 1.15, 4)
for (x, y, w, h) in cars:
    cv2.rectangle(image, (x, y), (x+w,y+h), color=(0, 255, 0), thickness=2)
plt.imshow(image)
```

```python
cv2.imshow('image', cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
